### Step 9. Which was the most-ordered item?

```python
most_ordered = df.groupby('item_name')['quantity'].sum().sort_values()
most_ordered.head(1)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
total_quantity = df.groupby('item_name')['quantity'].sum().max()
total_quantity
```

### Step 11. What was the most ordered item in the choice_description column?

```python
most_ordered = df.groupby('choice_description')['quantity'].sum().sort_values()
most_ordered.head(1)
```

### Step 12. How many items were orderd in total?

```python
total_quantity_choice = df.groupby('choice_description')['quantity'].sum().max()
total_quantity_choice
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
df['item_price'].dtype
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
df['item_price'] = df['item_price'].apply(lambda x: float(x.replace('$','')))
```

#### Step 13.c. Check the item price type

```python
df['item_price']
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue  = (df['item_price']*df['quantity']).sum
print(revenue)
```

### Step 15. How many orders were made in the period?

```python
total_orders = df['order_id'].nunique()
print(total_orders)
```

### Step 16. What is the average revenue amount per order?

```python
df['revenue'] = df['item_price']*df['quantity']
```

```python
avg_rev_perOrder = df.groupby('order_id')['revenue'].sum().mean()
print(avg_rev_perOrder)
```
